```python
import pandas as pd
import os
import numpy as np
import itertools
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
path=os.path.join('D:'+os.sep,'data Science','datascience IBM','data science with ml' ,'Projects','train.csv')
df=pd.read_csv(path)
```

```python
df.dropna(subset=["Survived"],axis=0,inplace=True)
```

```python
df[['Sex']]=df[['Sex']].replace(to_replace=['male','female'],value=[0,1])
df.head()
```

```python
cor=df.corr()
top_cor=cor.index
plt.figure(figsize=(15,15))
sns.heatmap(df[top_cor].corr(),annot=True,cmap="RdYlGn")
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=df,palette='RdBu_r')
```

```python
dataset=pd.get_dummies(df,columns=['Survived','Sex','SibSp','Parch'])
dataset.head()
```

```python
df.dtypes
```

```python
df[['Pclass']]=df[['Pclass']].astype(float)
df[['Fare']]=df[['Fare']].astype(float)
df.dtypes
```

```python
cols=['Pclass','Fare']
dataset[cols]=StandardScaler().fit_transform(dataset[cols])
dataset.head()
```

```python
y=df[['Survived']]
x=dataset[['Pclass']]
x=pd.concat([x,df[['Sex']]],axis=1)
#x=pd.concat([x,dataset[['Fare']]],axis=1)
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
x_train[:5]
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
k=21
mean_acc=np.zeros((k-1))
for i in range(1,k):
    neigh=KNeighborsClassifier(n_neighbors=i).fit(x,y)
    score=cross_val_score(neigh,x,y,cv=10)
    mean_acc[i-1]=score.mean()

print(mean_acc)
plt.plot(range(1,k),mean_acc,'g')
plt.xlabel("K Values")
plt.ylabel("Accuracy")
plt.show()
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion="entropy",max_depth=4)
tree.fit(x,y)
score1=cross_val_score(tree,x_train,y_train,cv=10)
print(score1.mean())
#Logistic Regression
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression(C=0.01).fit(x_train,y_train)
score2=cross_val_score(Lr,x_train,y_train,cv=10)
print(score2.mean())
#SVM
from sklearn import svm
clf=svm.SVC()
clf.fit(x_train,y_train)
score3=cross_val_score(clf,x_train,y_train,cv=10)
print(score3.mean())
#Random Forest
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier(max_depth=2, random_state=0)
clf1.fit(x_train,y_train)
score4=cross_val_score(clf1,x_train,y_train,cv=10)
print(score4.mean())
```
